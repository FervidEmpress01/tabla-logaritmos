```python
import math
x = math.log(2,10)
print(x)
```

```python
def raiz(x):
    resultado = x
    print(f"Número inicial: {round(resultado, 6)}")
    for i in range(1, 6):
        resultado = math.sqrt(resultado)
        print(f"Raíz {i}: {round(resultado, 6)}")
    return round(resultado, 6)
print(raiz(2))

log_2 = 32 * 0.021897 * 0.434294
print(f"valor aprox: {round(log_2, 6)}")
print(f"valor real: {math.log(2,10)}")
error = abs(math.log(2,10) - log_2) / math.log(2,10)
print(f"Error relativo: {round(error, 6)}")
```
